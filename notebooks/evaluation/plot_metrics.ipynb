```python
import matplotlib.pyplot as plt

# Suponha que voc tenha trs listas de valores
lista1 = [1, 3, 5, 7, 9, 11, 13]
lista2 = [2, 4, 6, 8, 10, 12, 14]
lista3 = [3, 6, 9, 12, 15, 18, 21]

# Plote os valores de cada lista em um nico grfico de linhas
plt.plot(lista1, label='Lista 1')
plt.plot(lista2, label='Lista 2')
plt.plot(lista3, label='Lista 3')

# Adicione um ttulo e rtulos aos eixos
plt.title('Grfico de Linhas com Trs Listas')
plt.xlabel('ndice')
plt.ylabel('Valores')

# Adicione uma legenda para identificar as linhas
plt.legend()

# Exiba o grfico
plt.show()
```

```python
import plotly.graph_objects as go

f1_scores = [0.63, 0.77, 0.83, 0.76, 0.56, 0.32]
precision = [0.91, 0.88, 0.79, 0.63, 0.39, 0.19]
recall = [0.48, 0.69, 0.88, 0.95, 0.97, 0.95]
x_labels = ['(120,80,70)', '(130,90,80)', '(140,100,90)', '(150,110,100)', '(160,120,110)', '(170,130,120)']

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=x_labels,
    y=precision,
    name='Precision',
    text=precision,
    mode='lines+markers+text',
    textposition='top center'

))

fig.add_trace(go.Scatter(
    x=x_labels,
    y=recall,
    name='Recall',
    text=recall,
    mode='lines+markers+text',
    textposition='top center'

))

fig.update_layout(
    title='Precision and Recall Through (RGB) Mean Threshold Variation ',
    xaxis_title='(RGB) mean threshold',
    yaxis_title='Metrics'
)
fig.show()
```
